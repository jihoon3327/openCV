```python
import sys
import numpy as np
import cv2
from PIL import Image as im
```

```python
src = cv2.imread('/Users/jungjihoon/Library/Mobile Documents/com~apple~CloudDocs/00_DataScience/민형기/lenna.bmp')
```

```python
face_classifier = cv2.CascadeClassifier('/Users/jungjihoon/Library/Mobile Documents/com~apple~CloudDocs/00_DataScience/민형기/haarcascade_frontalface_alt2.xml')
eye_classifier = cv2.CascadeClassifier('/Users/jungjihoon/Library/Mobile Documents/com~apple~CloudDocs/00_DataScience/민형기/haarcascade_eye.xml')
```

```python
faces = face_classifier.detectMultiScale(src)
```

```python
for (x1, y1, w1, h1) in faces:
    cv2.rectangle(src, (x1, y1), (x1 + w1, y1 + h1), (255, 0, 255), 2)
    
    faceROI = src[y1 : y1+h1//2, x1 : x1+w1]
    eyes = eye_classifier.detectMultiScale(faceROI)
    
    for (x2, y2, w2, h2) in eyes:
        center = (x2+w2//2, y2+h2//2)
        cv2.circle(faceROI, center, w2//2, (255, 0, 0), 2, cv2.LINE_AA)
```
